### Importar los módulos y configurar el driver de Selenium

```python
driver_path = "/Users/bcpcoelab/Downloads/chromedriver_2"
```

```python
# importar los módulos de la librería Selenium
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
```

```python
options = Options()
#options.add_argument("--headless") #
options.add_argument('--no-sandbox')
options.add_argument("--disable-extensions")
options.add_argument('start-maximized')
options.add_argument('disable-infobars')
```

```python
# Instanciar el objeto Driver usando los path definidos
driver = webdriver.Chrome(options=options, executable_path=driver_path)
```

```python
import json
```

```python
def get_row_data(row_tmp):
    row_head = ["TipoEstablecimiento", "FechaActualizacion", "Producto", "Laboratorio", "FarmaciaBotica", "PrecioUnitario"]
    id_list = ['Medicamento', 'Presentacion', 'MontoEmpaque', 'Pais', 'Registro', 'CondicionV', 'Marca', 'Titular', 'Fabricante', 'Label1', 'Estab', 'Direccion', 'Ubicacion', 'Telefono', 'Horario', 'DTecnico']
    
    row_col_tmp = row_tmp.find_elements_by_tag_name("td")
    row_values = [e.text for e in row_col_tmp[:6]]
    
    data_table = dict(zip(row_head, row_values))
    
    ver_mas = row_col_tmp[-1]
    ver_mas.click()
    
    window_list = driver.window_handles
    driver.switch_to.window(window_list[-1])
    
    data_dict = {}
    for id_ in id_list:
        data_dict[id_] = driver.find_element_by_id(id_).text
    
    driver.close()
    driver.switch_to.window(window_list[0])
    
    data_table.update(data_dict)
    
    print("-" * 100)
    print(json.dumps(data_table, indent=4), "\n")
    
    return data_table
```

```python
data_list = []
for row_tmp in row_list:
    data_list.append(get_row_data(row_tmp))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data_list)
```

```python
df.tail().T
```
