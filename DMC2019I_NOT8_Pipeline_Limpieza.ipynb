#### Limpieza

```python
import re
```

```python
df["product_price"] = df["product_price"].apply(lambda p: float(re.search(r'S/\s(.+)', p.strip()).group(1).replace(",", "")))
```

```python
df.loc[pd.notnull(df["product_discount"]), "product_discount"] = df.loc[pd.notnull(df["product_discount"]), "product_discount"].apply(lambda d: int(re.search(r'\d+', d).group()))
```
